```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
order_item = pd.read_csv("../../../downloads/olist_redefined/order_item.tsv", sep='\t').drop_duplicates()
order_item
```

```python
product = pd.read_csv("../../../downloads/olist_redefined/product.tsv", sep='\t').drop_duplicates()
product
```

```python
df = product.merge(order_item, on='product_id', how='left')
df = df.drop_duplicates().dropna()
df
```

```python
order_count = df.groupby(['seller_id', 'category'], as_index=False).agg({'order_id': 'count'})
order_count.rename(columns={"order_id":'order_count'}, inplace=True)
order_count.sort_values(['category', 'order_count', ], ascending=False).to_csv('re_tmp.csv', index=False)
# order_count.sort_values(['category', 'order_id', ], ascending=False)
```

```python
order_count['category'].unique()
```

```python
specific_categories = []

# specific_categories 리스트가 채워져 있으면 해당 카테고리만 선택하고,
# 비어있으면 모든 카테고리를 개수가 많은 순서대로 가져옵니다.
if specific_categories:
    all_categories = specific_categories
else:
    all_categories = order_count['category'].value_counts().index


# --- 3. Matplotlib Subplots으로 시각화 개선 ---
n_categories = len(all_categories)

# 선택된 카테고리 수에 따라 격자 크기를 유연하게 조절합니다.
if n_categories <= 3:
    ncols = n_categories
    nrows = 1
else:
    ncols = 3
    nrows = (n_categories + ncols - 1) // ncols

# 계산된 행/열에 맞춰 전체 그래프의 크기를 조절합니다.
figsize_width = 6 * ncols
figsize_height = 5 * nrows
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(figsize_width, figsize_height))

# 서브플롯이 하나일 경우에도 1차원 배열로 만듭니다.
if n_categories > 1:
    axes = axes.flatten()
else:
    axes = [axes]


# 각 카테고리에 대해 반복하여 서브플롯을 그립니다.
for i, category in enumerate(all_categories):
    ax = axes[i] # 현재 그릴 서브플롯 선택

    # 현재 카테고리에서 상위 10명의 판매자 데이터를 추출하고 주문량 순으로 정렬합니다.
    cat_data = order_count[order_count['category'] == category].head(10)
    cat_data = cat_data.sort_values('order_count', ascending=True)

    # 가로 막대그래프를 그립니다.
    bars = ax.barh(cat_data['seller_id'], cat_data['order_count'], color=sns.color_palette('plasma', n_colors=10))

    # 각 막대 끝에 주문량(값)을 텍스트로 표시합니다.
    ax.bar_label(bars, fmt='%d', padding=3, fontsize=9)

    # 각 서브플롯의 제목을 설정합니다.
    ax.set_title(f'Category: {category}', fontsize=12)
    ax.tick_params(axis='y', labelsize=8) # y축(seller_id) 폰트 크기 조절
    ax.set_xlabel('Order Count')

    # 텍스트가 잘리지 않도록 x축 범위를 조절합니다.
    if not cat_data.empty:
        ax.set_xlim(right=cat_data['order_count'].max() * 1.15)


# 사용하지 않는 나머지 서브플롯은 보이지 않게 처리합니다.
for i in range(len(all_categories), len(axes)):
    axes[i].set_visible(False)

# --- 4. 전체 제목 및 레이아웃 설정 ---
if specific_categories:
    if len(specific_categories) == 1:
        title_text = f"Top 10 Sellers in '{specific_categories[0]}' Category"
    else:
        title_text = 'Top 10 Sellers in Selected Categories'
else:
    title_text = 'Top 10 Sellers by Order Count in All Categories'

# y 파라미터로 제목 위치 조정 (기본값보다 살짝 아래로)
fig.suptitle(title_text, fontsize=20, y=0.95)

# 레이아웃 조정: 전체 제목 공간을 조금만 확보
plt.tight_layout(rect=[0, 0.03, 1, 0.95])


# --- 5. 그래프 출력 ---
plt.show()
```

# 판매 카테고리가 동일한 셀러 분포

```python
competitor_seller = order_count.groupby('category', as_index=False).agg({'seller_id':'count'})
competitor_seller.sort_values('seller_id', inplace=True, ascending=False)
```

```python
plt.figure(figsize=(15, 12))

# barplot을 사용하여 가로 막대 그래프를 생성합니다.
# y축에 카테고리 이름을, x축에 개수를 표시하여 차트를 회전합니다.
sns.barplot(
    x='seller_id',
    y='category',
    data=competitor_seller,
    palette='rocket', # 색상 테마 설정
    hue='category',
    legend=False
)

plt.title('Seller count Distribution by Product Category ', fontsize=16)
plt.xlabel('Number of seller (Count)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
# plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
