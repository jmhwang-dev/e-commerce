```python
import pandas as pd
import os
```

```python
file_path = r"../../../downloads/olist_redefined/order_status.tsv"
```

```python
df = pd.read_csv(file_path, sep='\t')
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# time_diff를 초(seconds) 단위로 변환 (NaT 제거)
diff_seconds = df['timestamp'].diff().dt.total_seconds()
diff_seconds.describe()
```

```python
quantiles = diff_seconds.quantile([0.5, 0.75, 0.9, 0.95, 0.99])
print(quantiles)
```
