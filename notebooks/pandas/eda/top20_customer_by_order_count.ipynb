```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
payment = pd.read_csv("../../../downloads/olist_redefined/payment.tsv", sep='\t').drop_duplicates()
df = payment['customer_id'].value_counts().to_frame()
df_counts = df.iloc[:20]
df_counts
```

```python
df_plot = df_counts.reset_index()
df_plot.columns = ['customer_id', 'count'] #   

#  20 .
df_plot = df_plot.head(20)


# --- 2.  ---
plt.figure(figsize=(12, 10))

# barplot     .
sns.barplot(
    x='count',          # x:  
    y='customer_id',    # y:  ID
    data=df_plot,
    palette='rocket',
    hue='customer_id',  #  : y  hue 
    legend=False        # (legend) 
)


# --- 3.      ---
plt.title('Top 20 Customers by Number of Payments', fontsize=16)
plt.xlabel('Number of Payments (Count)', fontsize=12)
plt.ylabel('Customer ID', fontsize=12)
plt.tight_layout()
```
