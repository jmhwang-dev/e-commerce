```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
order_item = pd.read_csv("../../../downloads/olist_redefined/order_item.tsv", sep='\t').drop_duplicates()
order_item
```

```python
product = pd.read_csv("../../../downloads/olist_redefined/product.tsv", sep='\t').drop_duplicates()
product
```

```python
order_product = product.merge(order_item, on='product_id', how='inner').drop_duplicates().dropna()
order_product = order_product.drop(columns='product_id')
order_product
```

```python
df = order_product['category'].value_counts().to_frame()
df
# order_product_value_counts
```

```python
plt.figure(figsize=(15, 12))

# barplot을 사용하여 가로 막대 그래프를 생성합니다.
# y축에 카테고리 이름을, x축에 개수를 표시하여 차트를 회전합니다.
sns.barplot(
    x='count',
    y='category',
    data=df,
    palette='rocket', # 색상 테마 설정
    hue='category',
    legend=False
)


# --- 3. 그래프 제목 및 레이블 설정 ---
plt.title('Product Category Distribution', fontsize=16)
plt.xlabel('Number of Products (Count)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
# plt.xticks(rotation=90)
# 레이아웃을 최적화하여 레이블이 잘리지 않게 합니다.
plt.tight_layout()


# --- 4. 그래프 출력 ---
# 생성된 그래프를 화면에 보여줍니다.
plt.show()
```

```python
total_count = df['count'].sum()
# 각 카테고리의 비율(%)을 계산하여 새로운 'percentage' 컬럼을 추가합니다.
df['percentage'] = (df['count'] / total_count) * 100

# 비율을 기준으로 데이터를 내림차순 정렬합니다.
df = df.sort_values(by='percentage', ascending=False)


# --- 3. Seaborn barplot으로 비율 시각화 ---
plt.figure(figsize=(15, 12))

# barplot을 사용하여 비율을 기준으로 가로 막대 그래프를 생성합니다.
ax = sns.barplot(
    x='percentage',
    y='category',
    data=df,
    palette='rocket',
    hue='category',
    legend=False
)


# --- 4. 그래프 제목, 레이블 및 주석 설정 ---
plt.title('Product Category Distribution (Percentage)', fontsize=16)
plt.xlabel('Percentage (%)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)

# 각 막대 옆에 비율(%) 텍스트 추가
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 0.1,  # 텍스트 x 위치
             p.get_y() + p.get_height() / 2, # 텍스트 y 위치
             f'{width:.3f}%', # 표시할 텍스트 (소수점 둘째 자리까지)
             ha='left',   # 수평 정렬
             va='center') # 수직 정렬

# x축의 범위를 넉넉하게 설정하여 텍스트가 잘리지 않게 합니다.
plt.xlim(0, df['percentage'].max() * 1.1)
plt.tight_layout()


# --- 5. 그래프 출력 ---
plt.show()
```
