```python
import pandas as pd
import os
```

```python
REDEFINED_DIR = "../../../downloads/olist_redefined"
STREAM_DST = os.path.join(REDEFINED_DIR, 'stream')
BATCH_DST = os.path.join(REDEFINED_DIR, 'batch')
os.makedirs(STREAM_DST, exist_ok=True)
os.makedirs(BATCH_DST, exist_ok=True)
```

`Orders`

```python
df = pd.read_csv("../../../downloads/olist/olist_orders_dataset.csv")
df.columns
```

```python
from pathlib import Path
customers = pd.read_csv(Path(BATCH_DST) / 'customers.tsv', sep='\t')
customer_ids = customers.drop(columns='zip_code')
df = pd.merge(df, customer_ids, on='customer_id', how='left')
df.drop(columns='customer_id', inplace=True)
df.rename(columns={"customer_unique_id": 'customer_id'}, inplace=True)
```

### estimated_delivery

```python
estimated_delivery = df[['order_estimated_delivery_date', 'order_id', 'customer_id']]
estimated_delivery = estimated_delivery.rename(columns={'order_estimated_delivery_date': 'estimated_delivery'})
estimated_delivery = estimated_delivery.sort_values('estimated_delivery')
estimated_delivery.to_csv(f"{STREAM_DST}/delivery_estimated.tsv", index=False, sep='\t')
```
