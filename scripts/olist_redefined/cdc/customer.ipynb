```python
import pandas as pd
import os
from pathlib import Path
```

```python
REDEFINED_DIR = "../../../downloads/olist_redefined"
STREAM_DST = Path(os.path.join(REDEFINED_DIR, 'stream'))
CDC_DST = Path(os.path.join(REDEFINED_DIR, 'cdc'))
os.makedirs(STREAM_DST, exist_ok=True)
os.makedirs(CDC_DST, exist_ok=True)
```

```python
customers = pd.read_csv("../../../downloads/olist/olist_customers_dataset.csv").drop_duplicates()
```

```python
customers = customers.rename(columns={'customer_zip_code_prefix':'zip_code', "customer_city":'city', 'customer_state':'state'})
customers = customers.drop(columns=['city', 'state'])
customers.sort_values(['customer_unique_id','zip_code'], inplace=True)
customers.to_csv(f"{CDC_DST}/customer.tsv", index=False, sep='\t')
```
