```python
import pandas as pd
import os
```

```python
from pathlib import Path

REDEFINED_DIR = "../../downloads/olist_redefined"
os.makedirs(REDEFINED_DIR, exist_ok=True)
```

```python
review_inference_arfifact = pd.read_csv("../../artifacts/postprocess/eng_reviews_with_senti.tsv", sep='\t')
review_inference_arfifact = review_inference_arfifact.rename(columns={'column_name': 'message_type', 'sentimentality': 'main_sentiment', 'por2eng': 'eng'})
review_inference_arfifact = review_inference_arfifact.drop_duplicates()
review_inference_arfifact.shape
```

```python
sentiment_score_df = pd.read_csv("../../artifacts/inference/sentiment_20250628_160638.tsv", sep='\t')
sentiment_score_df = sentiment_score_df.rename(columns={"por2eng": 'eng'}).drop_duplicates()
sentiment_score_df.shape
```

```python
review_inference = review_inference_arfifact.merge(sentiment_score_df, on='eng', how='inner')
review_inference = review_inference.drop_duplicates()
review_inference.head(1)
```

```python
review = pd.read_csv("../../downloads/olist_redefined/review.tsv", sep='\t')
review_metadata = review.drop(columns=['review_comment_title', 'review_comment_message'])
review_metadata.head(2)
```

```python
SILVERDEV_DIR = "../../downloads/silver_dev"
os.makedirs(SILVERDEV_DIR, exist_ok=True)

review_inference.to_csv(f"{SILVERDEV_DIR}/review_inference.tsv", sep='\t', index=False)
review_metadata.to_csv(f"{SILVERDEV_DIR}/review_metadata.tsv", sep='\t', index=False)
```
