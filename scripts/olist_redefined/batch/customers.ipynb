```python
import pandas as pd
import os
```

```python
REDEFINED_DIR = "../../../downloads/olist_redefined"
STREAM_DST = os.path.join(REDEFINED_DIR, 'stream')
BATCH_DST = os.path.join(REDEFINED_DIR, 'batch')
os.makedirs(STREAM_DST, exist_ok=True)
os.makedirs(BATCH_DST, exist_ok=True)
```

```python
customers = pd.read_csv("../../../downloads/olist/olist_customers_dataset.csv").drop_duplicates()
```

```python
customers = customers.rename(columns={'customer_zip_code_prefix':'zip_code', "customer_city":'city', 'customer_state':'state'})
customers.to_csv(f"{BATCH_DST}/customers.tsv", index=False, sep='\t')
```
