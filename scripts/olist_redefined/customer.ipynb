```python
import pandas as pd
import os
from pathlib import Path
```

```python
REDEFINED_DIR = "../../downloads/olist_redefined"
os.makedirs(REDEFINED_DIR, exist_ok=True)
```

```python
customer_origin = pd.read_csv("../../downloads/olist/olist_customers_dataset.csv").drop_duplicates()
```

```python
customer_df = customer_origin.rename(columns={'customer_zip_code_prefix':'zip_code'})
customer_df = customer_df.drop(columns=['customer_city', 'customer_state'])
customer_df = customer_df.sort_values(['customer_id','zip_code'])
customer_df.head(2)
```

```python
customers_cleaned = customer_df.drop(columns='customer_unique_id').drop_duplicates()
customers_cleaned.head()
```

```python
customers_cleaned.sort_values('zip_code').to_csv(f'{REDEFINED_DIR}/customer.tsv', sep='\t', index=False)
```
