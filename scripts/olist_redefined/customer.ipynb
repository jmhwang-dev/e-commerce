```python
import pandas as pd
import os
from pathlib import Path
```

```python
REDEFINED_DIR = "../../downloads/olist_redefined"
os.makedirs(REDEFINED_DIR, exist_ok=True)
```

```python
customer_origin = pd.read_csv("../../downloads/olist/olist_customers_dataset.csv").drop_duplicates()
```

```python
customer_origin = customer_origin.rename(columns={'customer_zip_code_prefix':'zip_code', "customer_city":'city', 'customer_state':'state'})
customer_origin = customer_origin.drop(columns=['city', 'state'])
customer_origin.sort_values(['customer_unique_id','zip_code'], inplace=True)
```

```python
customers_cleaned = customer_origin.drop(columns='customer_id').drop_duplicates()
customers_cleaned = customers_cleaned.rename(columns={"customer_unique_id": "customer_id"})
customers_cleaned.head()
customers_cleaned.sort_values('zip_code').to_csv(f'{REDEFINED_DIR}/customer.tsv', sep='\t', index=False)
```
